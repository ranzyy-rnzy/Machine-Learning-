2. **Столбчатая диаграмма для категориальной переменной**

```python

# Создаем столбчатую диаграмму
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', data=df,
              palette='pastel',
              order=df['Sex'].value_counts().index)

# Добавляем аннотации
for p in plt.gca().patches:
    plt.gca().annotate(f'{(p.get_height())}',
                      (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha='center', va='center', fontsize=10, color='black',
                      xytext=(0, 5), textcoords='offset points')

plt.title('Распределение по полу', fontsize=14)
plt.xlabel('Пол', fontsize=12)
plt.ylabel('Количество наблюдений', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

На графике представлено распределение по полу. Видно, что количество наблюдений для мужчин и женщин практически одинаково: 5001 для мужчин и 4999 для женщин. Это указывает на сбалансированное распределение по полу в выборке.

Если добавить категориальную переменную с помощью параметра hue, можно проанализировать, как распределение по полу связано с другими факторами.

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    x='Sex',
    y='Salary Per Month',
    hue='Occupation',
    data=df,
    ci=95  # доверительный интервал
)

plt.title('Распределение зарплат по полу и уровню занятости')
plt.xlabel('Пол')
plt.ylabel('Средняя зарплата (рупии)')
plt.legend(title='Уровень занятости')
plt.grid(True)
plt.show()
```

**Боксплот для числовой переменной:**

```python

# Создаем боксплот для зарплаты
plt.figure(figsize=(12, 8))
boxplot = sns.boxplot(
    y='Salary Per Month',
    data=df,
    palette='viridis',
    showfliers=True  # показывать выбросы
)

# Добавляем медиану красным цветом
sns.pointplot(
    y='Salary Per Month',
    data=df,
    color='red',
    ci=None
)

plt.title('Распределение зарплат сотрудников', fontsize=16)
plt.ylabel('Месячная зарплата (рупии)', fontsize=14)
plt.xlabel('')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Выбросы: На графике виден один выброс, отмеченный красной точкой, который находится значительно ниже основного диапазона зарплат. Это может указывать на аномально низкую зарплату, которая требует дополнительного анализа.

Медиана: Медиана зарплат находится примерно на уровне 85 000 рупий, что является центральной точкой распределения. Это значение делит данные на две равные части: половина сотрудников получает зарплату выше этой суммы, а другая половина — ниже.

Разброс данных: Межквартильный размах (IQR) показывает, что 50% сотрудников имеют зарплаты в диапазоне от около 80 000 до 100 000 рупий. Это говорит о том, что зарплаты в целом распределены достаточно равномерно в этом интервале.

```python

# Выберем только числовые столбцы
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Проверим, какие столбцы выбраны
print("Числовые столбцы:", numeric_cols)

# Создадим корреляционную матрицу
corr_matrix = df[numeric_cols].corr()

# Визуализация тепловой карты
plt.figure(figsize=(14, 10))
sns.heatmap(
    corr_matrix,
    annot=True,  # показывать значения корреляции
    cmap='coolwarm',  # цветовая схема
    fmt=".2f",  # формат чисел
    vmin=-1,  # минимальное значение
    vmax=1,  # максимальное значение
    square=True,  # квадратные ячейки
    linewidths=0.5  # толщина линий между ячейками
)
plt.title('Корреляционная матрица числовых переменных', fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
